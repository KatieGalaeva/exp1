```python
n, m = map(int, input().split())
image_coords=[]
for i in range(n):
    row=list(input().replace(' ', ''))   
    introw=[int(el) for el in row]      
    image_coords.append(introw)

def max_letter(matrix):
    length=0  
    for i in range(n):
        for j in range(m):
            a=b=c=d=0
            if matrix[i][j]==0:   
                continue
            cords=[i,j]
            next_x=cords[0]
            next_y=cords[1]
            #top
            while next_x>0:
                next_x-=1
                if (next_y < 0 or next_y > m-1) or (next_x < 0 or  next_x > n-1):
                    break
                elif matrix[next_x][next_y]==0:
                    break
                if matrix[next_x][next_y]==1:
                    a+=1
            next_x=cords[0]
            next_y=cords[1]
            #bottom
            while next_x<n-1:
                next_x+=1
                if (next_y < 0 or next_y > m-1) or (next_x < 0 or  next_x > n-1):
                    break
                elif matrix[next_x][next_y]==0:
                    break
                if matrix[next_x][next_y]==1:
                    b+=1
            next_x=cords[0]
            next_y=cords[1]
            #right
            while next_y<m-1:
                next_y+=1
                if (next_y < 0 or next_y > m-1) or (next_x < 0 or  next_x > n-1):
                    break
                elif matrix[next_x][next_y]==0:
                    break
                if matrix[next_x][next_y]==1:
                    c+=1
            next_x=cords[0]
            next_y=cords[1]
            #left
            while next_y>0:
                next_y-=1
                if (next_y < 0 or next_y > m-1) or (next_x < 0 or  next_x > n-1):
                    break
                elif matrix[next_x][next_y]==0:
                    break
                if matrix[next_x][next_y]==1:
                    d+=1
        
            print(a)
            print(b)
            print(c)
            print(d)
            if a==0 or b==0 or c==0 or d==0:
                cur_len=a+b+c+d+1
            length=max(cur_len,length)
    return length            
print(max_letter(image_coords))
```
