```python

def load():
    time = (input())
    time_split = time.split(':')
    start_time = int(time_split[0])*60+int(time_split[1])*1 +int(time_split[2])/60
    n = int(input())
    a=[]
    
    for i in range(n):
        row = list(input().split('"'))
        
        team='"'+row[1]+'"'
        other=row[2].split(' ')

        
        input_time=other[1]

        t=input_time.split(':')
        


        total_minutes= int(t[0])*60+int(t[1])*1 +int(t[2])/60
        
        a.append((team, total_minutes, other[2], other[3]))

    return start_time, n, a


start_time, n, a = load()
def search(start_time, n, a):
    d={}
    for i in range(n):
        min= 0
        num_server=0
        d[a[i][0]]=[num_server, min]
    t=20
    tem={}
    for i in range(n):
        tem[a[i][0]]={}

    for i in range(n):
        if a[i][3]=="PONG":
            continue

        if a[i][3]=="ACCESSED" and a[i][2] in tem[a[i][0]]:
            for j in tem[a[i][0]]:
                if j==a[i][2]:
                    extra=0
                    if a[i][1]<start_time:
                        extra = 1440
                    d[a[i][0]][1]= int(a[i][1] - start_time + extra)+tem[a[i][0]][j]*t
                    d[a[i][0]][0] += 1           
                
        if a[i][3]=="ACCESSED" and a[i][2] not in tem[a[i][0]]:
            extra=0
            if a[i][1]<start_time:
                extra = 1440
            d[a[i][0]][1]= int(a[i][1] - start_time + extra)
            d[a[i][0]][0] += 1

        if a[i][3]=="FORBIDEN" or a[i][3]=="DENIED" :
            tem[a[i][0]][a[i][2]] = 1

    final=[]
    for key, value in d.items():
        if value[0]>0:
            final.append((value[0], value[1], key))
    final_sort = sorted(final)    

    return final_sort
final = search(start_time, n, a)

n=len(final)
curr_place = 1
curr_ser = final[0][0]
curr_time = final[0][1]
for i in range(n):
    
    
    if curr_ser == final[i][0] and curr_time == final[i][1]:
        curr_ser = final[0][0]
        place = curr_place
        curr_time=final[i][1]
    else:
        place = i + 1
        curr_place = place
        curr_ser = final[i][0]
        curr_time = final[i][1]

    print(place, final[i][2], final[i][0], final[i][1])
     
```
