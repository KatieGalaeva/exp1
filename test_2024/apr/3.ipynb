```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.child = []


def binary_search(arr, x, key=None):
    low = 0
    high = len(arr) - 1
    mid = 0
 
    while low <= high:
        mid = (high + low) // 2
        cmp = key(arr[mid]) if key else arr[mid]
        if cmp < x:
            low = mid + 1
        elif cmp > x:
            high = mid - 1
        else:
            return mid
 
    return low

        
def buildTree(path_list):
    if not path_list:
        return None

    root = TreeNode(path_list[0][0])
    for path in path_list:
        cur_node = root
        for i in range(1, len(path)):
            token = path[i]
            bi_idx = binary_search(cur_node.child, token, key=lambda node: node.val)
            if bi_idx < len(cur_node.child) and cur_node.child[bi_idx].val == token:
                cur_node = cur_node.child[bi_idx]
            else:
                new_node = TreeNode(token)
                cur_node.child.insert(bi_idx, new_node)
                cur_node = new_node

    return root


def printTree(node, tab_str = ''):
    print(tab_str + node.val)
    for child in node.child:
        printTree(child, tab_str + '  ')


n=int(input())
a=[]
for i in range(n):
    list_root=list(input().split('/'))
    a.append(list_root)

tree = buildTree(a)
printTree(tree)
```
